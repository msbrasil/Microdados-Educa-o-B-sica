```python
matriculas_basico_total = int(censo_parana['QT_MAT_BAS'].sum())
print(f'Total de Matrículas do Ensino Básico: {matriculas_basico_total} (Infantil, Fundamental e Médio)')
basico_inf_fund = int(censo_parana[['QT_MAT_INF', 'QT_MAT_FUND']].sum().sum())
print(f'Quantidade de matrículas do ensino infantil e fundamental: {basico_inf_fund}')
perc_inf_fund_total = basico_inf_fund / matriculas_basico_total
print(f'Percentual de matrículas do infantil e fumdamental no total: {perc_inf_fund_total*100:.2f}%')
```

```python
professores_basico_total = int(censo_parana['QT_DOC_BAS'].sum())
print(f'Total de Matrículas do Ensino Básico: {professores_basico_total} (Infantil, Fundamental e Médio)')
professores_inf_fund = int(censo_parana[['QT_DOC_INF', 'QT_DOC_FUND']].sum().sum())
print(f'Quantidade de matrículas do ensino infantil e fundamental: {professores_inf_fund}')
profs_perc_inf_fund_total = professores_inf_fund / professores_basico_total
print(f'Percentual de matrículas do infantil e fumdamental no total: {profs_perc_inf_fund_total*100:.2f}%')
```
